```python
import os
import imageio.v3 as iio
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img_dir = os.path.join("C:", os.sep, "Users", "jeos", "Prj","WB","DST","DS","data","raw")
img_path1 = os.path.join(img_dir,"Apple___Apple_scab", "image (1).JPG")
img_path2 = os.path.join(img_dir,"Cleavers", "1.png")
```

```python
image1=iio.imread(img_path1)
image2=iio.imread(img_path2)

plt.figure(figsize=(10,4))

plt.subplot(121)
plt.imshow(image1);

plt.subplot(122)
plt.imshow(image2);
```

```python
print(image1.min(),image1.mean(),image1.max())
print(image2.min(),image2.mean(),image2.max())
```

```python
print(image1.shape)
print(image2.shape)
```

```python
plt.figure(figsize=(10,4))
for chn in range(image1.shape[2]):
  plt.subplot(1,3,chn+1)
  plt.imshow(image1[:,:,chn], cmap="gray");
```

```python
plt.figure(figsize=(10,4))
for chn in range(image2.shape[2]):
  plt.subplot(1,3,chn+1)
  plt.imshow(image2[:,:,chn], cmap="Accent");
```

```python
print(type(image1), image1.min(),int(np.round(image1.mean(),0)),image1.max(), image1.sum(), sep = "\t")
for chn in range(image1.shape[2]):
  print(type(image1[:,:,chn]),int(np.round(image1[:,:,chn].mean())), image1[:,:,chn].sum(), sep = 2*"\t")
  print(np.percentile(image1[:,:,chn],[0,25,50,75,100]))
  
```

```python
print(type(image2), image2.min(),int(np.round(image2.mean(),0)),image2.max(), image2.sum(), sep = "\t")
for chn in range(image2.shape[2]):
  print(type(image2[:,:,chn]),int(np.round(image2[:,:,chn].mean())), image2[:,:,chn].sum(), sep = 2*"\t")
  print(np.percentile(image2[:,:,chn],[0,25,50,75,100]))
```
