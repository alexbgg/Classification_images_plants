```python
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

my_path = "../data/1.0-bbri-dataframe.csv"
path_save = "../reports/figures/"

# Creating the dataframe
df = pd.read_csv(my_path)

# Show first records
display(df.head())
```

```python
# Just execute this cell to create some plot routines

figsize = (10, 10)

sns.set_theme()

# Plot routines

def do_plot_countbyformat(num : int):
  plt.figure(figsize = figsize)
  plt.title('Plot ' + f"{num}" + ' - Count of pictures by format')
  plt.ylabel('format')
  plt.xlabel('count')
  sns.countplot(y = df.Image_format)
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()

def do_plot_countbymode(num: int):
  plt.figure(figsize = figsize)
  plt.title('Plot ' + f"{num}" + ' - Count of pictures by mode')
  plt.ylabel('format')
  plt.xlabel('count')
  sns.countplot(y = df.Image_mode)
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()

def do_plot_boxplotbyshape(num: int):
  d1 = {'ax_type' : ['width'] * len(df.Image_width), 'px' : df.Image_width}
  d2 = {'ax_type' : ['height'] * len(df.Image_height), 'px' : df.Image_height}
  df2 = pd.concat([pd.DataFrame(d1), pd.DataFrame(d2)], axis = 0)

  plt.figure(figsize = figsize)
  sns.boxplot(x = df2.ax_type, y = df2.px)
  plt.title('Plot ' + f"{num}" + ' - Boxplot on Image dimensions')
  plt.xlabel('')
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()

def do_plot_countbyplant(num: int):
  plt.figure(figsize = figsize)
  plt.title('Plot ' + f"{num}" + ' - Count of pictures by plant')
  plt.ylabel('plant')
  plt.xlabel('count')
  plt.yticks(fontsize = 'x-small', rotation = 30)
  sns.countplot(y = df.Plant.sort_values())
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()

def do_plot_countbyillness(num: int):
  plt.figure(figsize = figsize)
  plt.title('Plot ' + f"{num}" + ' - Count of pictures by disease')
  plt.ylabel('disease')
  plt.xlabel('count')
  plt.yticks(fontsize = 'x-small', rotation = 30)
  sns.countplot(y = df.Illness.sort_values())
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()

def do_plot_piebyformat(num : int):
  plt.figure(figsize = figsize)
  plt.title('Plot ' + f"{num}" + ' - Image extensions')

  format_count = df.Image_format.value_counts()
  x = format_count.values
  labels = format_count.index
  total = np.sum(x)

  plt.pie(x = x, labels = labels, autopct = lambda x: str(round(x*total/100)))
  plt.legend()
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()

def do_plot_piebymode(num : int):
  plt.figure(figsize = figsize)
  plt.title('Plot ' + f"{num}" + ' - Mode of images')

  mode_count = df.Image_mode.value_counts()
  x = mode_count.values
  labels = mode_count.index
  total = np.sum(x)

  plt.pie(x = x, labels = labels, autopct = lambda x: str(round(x*total/100)))
  plt.legend()
  plt.savefig(path_save + '1-bbri-plot-' + f"{num:02d}" + '.png')
  plt.show()
  plt.close()
```

```python
# Display plot, it's automatically also saved to the folder
do_plot_countbyformat(1)
```
