```python
import cv2 #import OpenCV
import matplotlib.pyplot as plt
import pandas as pd
import os

print(os.getcwd())

my_path = "../data/1.0-bbri-dataframe.csv"

# Creating the dataframe
df = pd.read_csv(my_path, index_col = 0)

# Filter to temporarily reduce magnitude of data during evaluation period
df = df [(df.Plant == "Sugar beet") | (df.Plant == "Strawberry") ]

# Show first records
display(df.head())
```

```python
# Defining target format
format_x = 128
format_y = 128

path_read = "../data/raw/"
path_save = "../data/processed/"
path_save = "G:/Meine Ablage/Colab Notebooks/data/processed/"

# Iterating through our dataframe
for index, row in df.iterrows():

# Getting the attributes
  dirname = row.Image_folder + "/"
  filename = row.Image_name
  Image_width = row.Image_width
  Image_height = row.Image_height

# Reading source picture
  to_read = path_read + dirname + filename
  img = cv2.imread(to_read, cv2.IMREAD_COLOR)

# Adjusting shape if required
  if (Image_width != format_x) & (Image_height != format_y):
    if Image_width * Image_height < format_x * format_y:
      interpolation = cv2.INTER_CUBIC      # Scale up
    else:
      interpolation = cv2.INTER_AREA       # Scale down
    img = cv2.resize(img, dsize = (format_x, format_y), interpolation = interpolation)

# Creating the target directory if it doesn't exist yet
  to_write = path_save + dirname
  if not(os.path.isdir(to_write)):
    os.mkdir(to_write)

# Saving the target file
  to_write = to_write + filename
  is_written = cv2.imwrite(to_write, img)

  if (is_written):
    print(to_write, "ok")
  else:
    print(to_write, "nok")
```
